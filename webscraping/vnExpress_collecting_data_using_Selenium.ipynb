```python
# import libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

from bs4 import BeautifulSoup
```

```python
# Đường dẫn đến Chrome WebDriver trên máy tính của bạn
chromedriver_path = r"C:\3.chromedriver-win32\chromedriver-win32\chromedriver.exe"

# Khai báo đường dẫn cho Chrome WebDriver bằng cách thiết lập biến môi trường
webdriver.chrome.driver = chromedriver_path

# Khởi tạo trình duyệt Chrome
driver = webdriver.Chrome()
```

```python
# https://www.youtube.com/watch?v=PPk5lP5nFfE&list=PLYlCX2e2RsgnPZXCqonA2nrKkGuNoQY-p&index=1
# Mở trang web vnexpress
driver.get("https://vnexpress.net/")
```

```python
print(driver.title)
```

```python
print(driver.page_source)
```

```python
# Duyệt qua từng bài viết và in tiêu đề
articles = driver.find_elements(By.CSS_SELECTOR,'article.item-news')
for article in articles:
    try:
        title_elements = article.find_elements(By.TAG_NAME, 'h3')
        for title_element in title_elements:
            title = title_element.text
        desc =  article.find_element(By.TAG_NAME, 'p').text
        link =  article.find_element(By.CSS_SELECTOR,'h3.title-news > a').get_attribute('href')

    except NoSuchElementException:
        title = "N/A"
        desc = "N/A"
        link = "N/A"

    print(title)
    print(link)
    print(desc)
    print('----------------')
```

```python
# đóng trang web
driver.close()
```
